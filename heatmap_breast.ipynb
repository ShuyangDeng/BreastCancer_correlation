```python
import os
import pandas as pd
pro = pd.read_csv('/Users/dengshuyang/Desktop/Bioinfo2019/Breast_project/Protein_match2.csv',index_col='Unnamed: 0')
rna = pd.read_csv('/Users/dengshuyang/Desktop/Bioinfo2019/Breast_project/RNA_match2.csv',index_col='Unnamed: 0')
cna = pd.read_csv('/Users/dengshuyang/Desktop/Bioinfo2019/Breast_project/CNA_match2.csv',index_col='Unnamed: 0')
```

```python
df_pam50=pd.read_csv('/Users/dengshuyang/Desktop/Bioinfo2019/Breast_project/pam50_genes.csv')
pam50=sorted(df_pam50['gene'].values)
print(pam50)
```

```python
pro = pro[rna.columns]
pro
```

```python
rna
```

```python
import numpy as np
rna_cna_spearman=[]
for gene_rna in pam50:
    if gene_rna in rna.index:
        
        for gene_cna in pam50:
            if gene_cna in cna.index:
                #data = gene_rna,gene_cna,np.corrcoef(rna.ix[gene_rna],cna.ix[gene_cna])[0][1] # same order
                data = gene_rna, gene_cna,rna.ix[gene_rna].corr(cna.ix[gene_cna], method="spearman")
                rna_cna_spearman.append(data)
# also make different color for positive and negtive
```

```python
pro_rna_spearman=[]
for gene_pro in pam50:
    if gene_pro in pro.index:

        for gene_rna in pam50:
            if gene_rna in rna.index:
                #data = gene_pro,gene_rna,np.corrcoef(pro.ix[gene_pro],rna.ix[gene_rna])[0][1] # same order
                data = gene_pro, gene_rna,pro.ix[gene_pro].corr(rna.ix[gene_rna], method="spearman")
                print(data)
                pro_rna_spearman.append(data)
# also make different color for positive and negtive
```
